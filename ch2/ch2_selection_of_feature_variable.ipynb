```python
from sklearn.datasets import load_digits
from sklearn.feature_selection import SelectKBest, chi2, f_classif, mutual_info_classif

x, y = load_digits(return_X_y=True)
x.shape
```

```python
import numpy as np
print(x[:2])
```

```python
print(np.unique(y))
```

```python
sb = SelectKBest(score_func=chi2, k=20)
x_new = sb.fit_transform(x, y)
x_new.shape
```

```python
# get_support() : 어떤 특성변수가 선택(True)되었는지 파악

sb.get_support()
```

```python
# scores_ : 각 특성변수의 x^2 값

sb.scores_
```

```python
# pvalues_ : 각 특성변수의 p-value 출력

sb.pvalues_
```

### 선택된 30개의 특성변수를 PCA를 통해 3개의 차원으로 축소하고 이를 3D 시각화를 통해 살펴보기

```python
!pip install plotly
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.decomposition import PCA
import pandas as pd
import plotly.express as px
```

```python
md1 = make_pipeline(SelectKBest(k=30), PCA(n_components=3))
x_new = md1.fit_transform(x, y)
```

```python
# 3개로 차원축소된 특성변수를 DataFrame으로 pandas 데이터를 만들고 있음.

df1 = pd.DataFrame({'X': x_new[:, 0], 'Y':x_new[:, 1], 'Z':x_new[:, 2], 'C': y})
df1.head()
```

```python
# Plotly.express를 이용한 3D 시각화
# 3개의 특성변수와 클래스(color = 'C'로 지정)를 보여주고 있음.

# 10개의 숫자범주를 3개의 특성변수로 구별하기에는 문제가 있어 보임.

fig = px.scatter_3d(df1, x='X', y='Y', z='Z', color='C')
fig.show()
```

```python
# 비선형 PCA 
# kernelPCA를 이용하여 30개의 특성변수를 6개의 특성변수로 차원 축소
# 이 중 첫 3개의 PCA를 이용하여 그림을 그리면 숫자 1과 숫자 6을 쉽게 구분할 수 있음.

from sklearn.pipeline import make_pipeline
from sklearn.decomposition import KernelPCA
import pandas as pd
import plotly.express as px

md2 = make_pipeline(SelectKBest(k=30), KernelPCA(kernel='rbf', n_components=6))
x_new2 = md2.fit_transform(x, y)
```

```python
df2 = pd.DataFrame({'X': x_new2[:, 0], 'Y': x_new2[:, 1], 'Z': x_new2[:, 2], 'C' : y})
df2.head()
```

```python
fig = px.scatter_3d(df2, x='X', y='Y', z='Z', color='C')
fig.show()
```

```python
# 6개의 특성변수 중 나머지 3개의 축으로 그림을 그린다
# -> 0, 1, 2를 구별할 수 있는 특성변수임을 보여주고 있음.

# PCA를 통해 분류를 쉽게 할 수 있는 새로운 특성변수를 만들 수 있다.

df3 = pd.DataFrame({'X': x_new2[:, 3], 'Y': x_new2[:, 4], 'Z': x_new2[:, 5], 'C':y})
fig = px.scatter_3d(df3, x='X', y='Y', z='Z', color='C')
fig.show()
```

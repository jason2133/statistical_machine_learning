```python
from sklearn.datasets import make_circles
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X, y = make_circles(factor = 0.8, noise=0.2, random_state=3)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.colorbar()
plt.show()
```

```python
import hdbscan

def hdbscan2(num):
    hdb = hdbscan.HDBSCAN(min_samples=num)
    y_hdb = hdb.fit_predict(X)

    print('min_samples = %d' % (num))
    
    plt.scatter(x=X[y_hdb==0, 0], y=X[y_hdb==0, 1], marker='x', color='r', label='cluster1')
    plt.scatter(x=X[y_hdb==1, 0], y=X[y_hdb==1, 1], marker='o', color='b', label='cluster2')
    plt.legend()
    plt.show()
```

```python
hdbscan2(10)
```

```python
for i in range(1, 31):
    hdbscan2(i)
```

```python
from sklearn.datasets import make_circles
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X, y = make_circles(factor = 0.8, noise=0.2, random_state=3)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.colorbar()
plt.show()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2, init='k-means++',
           n_init=10, max_iter=300,
           tol=1e-04, random_state=3)
y_km = km.fit_predict(X)
```

```python
distortions = []

for i in range(1, 11):
    km = KMeans(n_clusters=i, n_init=10, max_iter=300, random_state=3)
    km.fit(X)
    distortions.append(km.inertia_)
```

```python
plt.scatter(x=np.arange(1, 11), y=distortions)
plt.plot(np.arange(1, 11), distortions)
plt.show()
```

```python
plt.scatter(x=X[y_km==0, 0], y=X[y_km==0, 1], marker='x', color='r', label='cluster1')
plt.scatter(x=X[y_km==1, 0], y=X[y_km==1, 1], marker='o', color='b', label='cluster2')
plt.scatter(x=km.cluster_centers_[:, 0], y=km.cluster_centers_[:, 1], marker='D', color='g', label='centroid')
plt.legend()
plt.show()
```

```python
km6 = KMeans(n_clusters=6, n_init=10, max_iter=300, random_state=3)
km6.fit(X)
y_km6 = km6.fit_predict(X)

plt.scatter(x=X[y_km6==0, 0], y=X[y_km6==0, 1], marker='x', color='r', label='cluster1')
plt.scatter(x=X[y_km6==1, 0], y=X[y_km6==1, 1], marker='o', color='b', label='cluster2')
plt.scatter(x=km6.cluster_centers_[:, 0], y=km6.cluster_centers_[:, 1], marker='D', color='g', label='centroid')
plt.legend()
plt.show()
```

```python
def kmeansplus(num):
    km = KMeans(n_clusters=num, n_init=10, max_iter=300, random_state=3)
    km.fit(X)
    y_km = km.fit_predict(X)
    print('n_clusters = %d' % (num))
    plt.scatter(x=X[y_km==0, 0], y=X[y_km==0, 1], marker='x', color='r', label='cluster1')
    plt.scatter(x=X[y_km==1, 0], y=X[y_km==1, 1], marker='o', color='b', label='cluster2')
    plt.scatter(x=km.cluster_centers_[:, 0], y=km.cluster_centers_[:, 1], marker='D', color='g', label='centroid')
    plt.legend()
    plt.show() 
```

```python
for i in range(1, 12):
    kmeansplus(i)
```

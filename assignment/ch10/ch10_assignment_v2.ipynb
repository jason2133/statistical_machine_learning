### SVM

```python
import pandas as pd
dat_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
```

```python
from sklearn.model_selection import train_test_split
X, y = dat_wine.iloc[:, 1:].values, dat_wine.iloc[:, 0].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline

pipe_svc = make_pipeline(StandardScaler(), SVC(random_state=1))
param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]
param_grid = [{'svc__C':param_range, 'svc__kernel':['linear']},
             {'svc__C':param_range, 'svc__gamma':param_range, 'svc__kernel':['rbf']}]
gs2 = GridSearchCV(estimator=pipe_svc, param_grid=param_grid, scoring='accuracy', cv=10)
gs2 = gs2.fit(X_train, y_train)
```

```python
print(gs2.best_score_)
```

```python
print(gs2.best_params_)
```

```python
clf2 = gs2.best_estimator_
clf2.fit(X_train, y_train)
print(clf2.score(X_test, y_test))
```

```python
print(clf2)
```

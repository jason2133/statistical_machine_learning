### 1. 188쪽의 wine_data를 이용하여 MDS, local MDS, LLE, Isomap, t-SNE의 성능을 비교하라

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.colors import ListedColormap
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
dat_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header=None)
```

```python
from sklearn.model_selection import train_test_split
X, y = dat_wine.iloc[:, 1:].values, dat_wine.iloc[:, 0].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_train_std = std.fit_transform(X_train)
X_test_std = std.transform(X_test)
```

```python
pca = PCA(n_components=None)
X_train_pca = pca.fit_transform(X_train_std)
print(pca.explained_variance_ratio_)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA

lr = LogisticRegression()
lpca = PCA(n_components=4)

X_train_pca = lpca.fit_transform(X_train_std)
X_test_pca = lpca.transform(X_test_std)
lr.fit(X_train_pca, y_train)

y_train_pre = lr.predict(X_train_pca)
y_test_pre = lr.predict(X_test_pca)

from sklearn import metrics
print(metrics.accuracy_score(y_train, y_train_pre))
print(metrics.accuracy_score(y_test, y_test_pre))
```

### MDS

```python
from sklearn.linear_model import LogisticRegression
from sklearn.manifold import MDS

lr = LogisticRegression()
model_mds = MDS(n_components=3, random_state=17)

X_train_mds = model_mds.fit_transform(X_train_std)
X_test_mds = model_mds.fit_transform(X_test_std)
lr.fit(X_train_mds, y_train)

y_train_pre_mds = lr.predict(X_train_mds)
y_test_pre_mds = lr.predict(X_test_mds)

print(metrics.accuracy_score(y_train, y_train_pre_mds))
print(metrics.accuracy_score(y_test, y_test_pre_mds))
```

```python
# MDS 초모수 조절

mds_train_accuracy = []
mds_test_accuracy = []

def mds_num(num):
    lr = LogisticRegression()
    model_mds = MDS(n_components=num, random_state=17)

    X_train_mds = model_mds.fit_transform(X_train_std)
    X_test_mds = model_mds.fit_transform(X_test_std)
    lr.fit(X_train_mds, y_train)

    y_train_pre_mds = lr.predict(X_train_mds)
    y_test_pre_mds = lr.predict(X_test_mds)
    
    mds_train_accuracy.append(metrics.accuracy_score(y_train, y_train_pre_mds))
    mds_test_accuracy.append(metrics.accuracy_score(y_test, y_test_pre_mds))
    
    print('n_components = %d' % (num))
    print(metrics.accuracy_score(y_train, y_train_pre_mds))
    print(metrics.accuracy_score(y_test, y_test_pre_mds))
    print('-' * 50)
```

```python
for i in range(1, 20):
    mds_num(i)
```

```python
import numpy as np
print('TRAIN')
mds_train_accuracy = np.array(mds_train_accuracy)
print(np.argmax(mds_train_accuracy))
print(mds_train_accuracy[np.argmax(mds_train_accuracy)])

print('-' * 50)

print('TEST')
mds_test_accuracy = np.array(mds_test_accuracy)
print(np.argmax(mds_test_accuracy))
print(mds_test_accuracy[np.argmax(mds_test_accuracy)])
```

### LLE

```python
from sklearn.linear_model import LogisticRegression
from sklearn.manifold import LocallyLinearEmbedding as LLE

lr = LogisticRegression()
model_lle = LLE(n_neighbors=40, n_components=3, random_state=17)

X_train_lle = model_lle.fit_transform(X_train_std)
X_test_lle = model_lle.transform(X_test_std)
lr.fit(X_train_lle, y_train)

y_train_pre_lle = lr.predict(X_train_lle)
y_test_pre_lle = lr.predict(X_test_lle)

print(metrics.accuracy_score(y_train, y_train_pre_lle))
print(metrics.accuracy_score(y_test, y_test_pre_lle))
```

```python
# LLE 초모수 조절

lle_train_accuracy = []
lle_test_accuracy = []

def lle_num(a, b):
    lr = LogisticRegression()
    model_lle = LLE(n_neighbors=a, n_components=b, random_state=17)

    X_train_lle = model_lle.fit_transform(X_train_std)
    X_test_lle = model_lle.fit_transform(X_test_std)
    lr.fit(X_train_lle, y_train)

    y_train_pre_lle = lr.predict(X_train_lle)
    y_test_pre_lle = lr.predict(X_test_lle)
    
    lle_train_accuracy.append(metrics.accuracy_score(y_train, y_train_pre_lle))
    lle_test_accuracy.append(metrics.accuracy_score(y_test, y_test_pre_lle))

    print('n_neighbors = %d, n_components = %d' % (a, b))
    print(metrics.accuracy_score(y_train, y_train_pre_lle))
    print(metrics.accuracy_score(y_test, y_test_pre_lle))
    print('-' * 50)
```

```python
for i in range(1, 51):
    for j in range(1, 14):
        lle_num(i, j)
```

```python
import numpy as np
print('TRAIN')
lle_train_accuracy = np.array(lle_train_accuracy)
print(np.argmax(lle_train_accuracy))
print(lle_train_accuracy[np.argmax(lle_train_accuracy)])

print('-' * 50)

print('TEST')
lle_test_accuracy = np.array(lle_test_accuracy)
print(np.argmax(lle_test_accuracy))
print(lle_test_accuracy[np.argmax(lle_test_accuracy)])
```

### Isomap
